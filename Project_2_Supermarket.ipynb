```python
import pandas as pd

# Sample Data
data = {
    'Store': ['Store A', 'Store A', 'Store B', 'Store B', 'Store C', 'Store C'],
    'Product': ['Product 1', 'Product 2', 'Product 1', 'Product 2', 'Product 1', 'Product 2'],
    'Sales': [100, 150, 200, 250, 300, 350]
}

# Create DataFrame
df = pd.DataFrame(data)
df
```

```python
#Group by 'Store' and calculate total sales for each store
grouped_df = df.groupby('Store')['Sales'].sum().reset_index()

#Print the result
print(grouped_df)
```

```python
#LAMBDA FUNCTION
```

```python
#Traditional function to add 10 to a number
def add_ten(x):
    return x + 10
print(add_ten(5)) #Output: 15
```

```python
#Equivalent lambda function
y = lambda x: x + 10
print(y(5)) #Output:15
```

```python
#Lambda Function to multiply two numbers
multiply = lambda x,y: x*y
#Using the lambda function
result = multiply(3,4)
print(result) #Output: 12 
```
