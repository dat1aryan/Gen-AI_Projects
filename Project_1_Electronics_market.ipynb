Question 2 :- Prepare the data frame for further analysis.

```python
#since there are 373700 products 372610 are non-null count products so there
```

```python
#Data Cleaning
#Since there may be a chance of duplicate data and null value.
df.isna().sum() #to check the total null value
```

```python
#To drop the null values:
df.dropna(inplace=True)
df.isna().sum()
```

```python
#Checking for duplicate values
df.duplicated().value_counts()
```

```python
dfa=df.drop_duplicates(subset=None , keep='first', inplace=False) # dropping the duplicate values
dfa
```

```python
dfa.columns = ['Order_ID', 'Product', 'Quantity_Ordered', 'Price_Each', 'Order_Date', 'Purchase_Address']
dfa.head()
```

```python
dfa['Date'] = dfa['Order_Date'].str[:8] #making a seperate column for Date
dfa
```

```python
dfa['Month'] = dfa['Order_Date'].str[:2] #making a seperate column for month
dfa
```

```python
dfa[['House_No.','City','Pin_Code']] = dfa.Purchase_Address.str.split(",",expand=True)
dfa
```

```python
#Check for month column for bad data entry and remove it - concept is to remove bad data entry
dfa.Month.value_counts() #to check the unique values #stucked here
```

```python
#Here we can see that "Or" value which is not a month so we are going to remove these rows which has these bad data.
```

```python
dfa= dfa[dfa['Month'].str[:2] !='Or']
dfa.head()
```

```python
#Correct data types
dfa['Month'] = dfa['Month'].astype('int64')
dfa['Order_Date'] = pd.to_datetime(dfa['Order_Date'])
dfa['Date'] = pd.to_datetime(dfa['Date'])
dfa.info()
```

```python
df.info()
```

```python
dfa['Quantity_Ordered'] = dfa['Quantity_Ordered'].astype('int32')
dfa['Price_Each'] = dfa['Price_Each'].astype('float64')
dfa.info()
```

```python
#Sales column - 
dfa['Sales'] = dfa['Quantity_Ordered'] * dfa['Price_Each']
dfa
```

```python
#Sales Values
Monthly_sales = dfa.groupby('Month').Sales.sum()
Monthly_sales
```

```python
 # Summary for above steps:
# 1 In new data set we have seperate column for Time,date,Month,House_No....
# 2 Sales column
# 3 In data set all null values are eliminated
# 4 In data set order_id is repeating so we cannot keep it as index number
# 5 In old data set we had some null values, or in months all are cleared
# 6 Duplicate Data is removed from old data set
```

```python
# Which city had the highest sales overall?
city_sales=dfa.City.value_counts()
city_sales
```

```python
plt.figure(figsize=(16,6))
sns.countplot(x=dfa['City'])
plt.xlabel("City Name")
plt.ylabel("Number of Sells")
plt.title("City vs Sales")
```

```python
# Distribution of sales over different times of day ( Morning, Afternoon, Evening, Night)
dfa['Hour'] = dfa.Order_Date.dt.hour
```

```python
dfa.groupby('Hour').count()
```

```python
hours = [hour for hour, dfa in dfa.groupby('Hour')]
plt.plot(hours, dfa.groupby('Hour').count())
plt.grid(True)
plt.xticks(hours)
plt.xlabel('Hours in 24 Hr Format')
plt.ylabel('Orders')
plt.show()
```

```python
#7 pm most orders placed because people like shopping after office
#3 am - 4 am is least because most people must be sleeping
#11am - 12 noon is maximum after 7 pm
#store can come with a offer and discount in the night and early morning to make the sales go high in those times too.
```

```python
# Highest product sold by amount and numbers
```

```python
#Highest Orders and Highest Sales
Highest_orders = dfa.Product.value_counts()
Highest_sales = dfa.groupby('Product').Sales.sum().sort_values(ascending=False)

#plotting both the charts
fig, axes = plt.subplots(1, 2, figsize=(20,7))

sns.barplot(ax=axes[0], x=Highest_orders.index, y=Highest_orders.values)
axes[0].set_xlabel('Products')
axes[0].set_ylabel('No. of Orders')
axes[0].tick_params(axis='x' , rotation=90)

sns.barplot(ax=axes[1],x=Highest_sales.index, y=Highest_sales.values)
axes[1].set_xlabel('Products')
axes[1].set_ylabel('Sales Amount')
axes[1].tick_params(axis='x', rotation=90)
plt.show()
```
