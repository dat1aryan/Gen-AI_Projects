# Analyzing Outliers


```python
# 1. Calculate the Pearson correlation coefficient (for relationship context)
correlation = df['Population'].corr(df['Total Cases'])

# 2. Identify top countries by Population (potential vertical outliers)
outliers_population = df.nlargest(3, 'Population')[['Country', 'Population', 'Total Cases']]

# 3. Identify top countries by Total Cases (potential horizontal outliers)
outliers_cases = df.nlargest(3, 'Total Cases')[['Country', 'Population', 'Total Cases']]

# 4. Combine unique countries for a comprehensive outlier table
outlier_data = pd.concat([outliers_population, outliers_cases]).drop_duplicates(subset=['Country'])

print(f"Pearson Correlation Coefficient (Population vs Total Cases): {correlation:.2f}")
print("\nData for Potential Outliers (Countries with Extreme Values):")
print(outlier_data)
```

This indicates a moderate positive linear relationship. Generally, countries with a higher
population tend to have a higher total number of cases, but the relationship is not strong.
This suggests that population size is only one factor influencing total case counts; other
factors (like policy, testing rates, and time) are also highly significant

---------------------------------------------------------------------------------------------------------------------------------------------------------
